### Distancia Euclidiana (Numpy)

```
import numpy as np
import matplotlib as plt
E1=np.array([10000,1,0,0,0,0,7,15,1])
E2=np.array([20000,0,1,1,0,1,3,3,0])
dist3=np.sqrt(np.sum((E1-E2)**2))
dist3
```

### Distancia Euclindiana (Scipy) 

```
from scipy.spatial import distance
E1=np.array([10000,1,0,0,0,0,7,15,1])
E2=np.array([20000,0,1,1,0,1,3,3,0])
dist3= distance.euclidean(E1,E2)
dist3
```

#### Eucludian Method


```
from scipy.spatial import distance
E1=np.array([10000,1,0,0,0,0,7,15,1])
E3=np.array([15000,1,1,2,1,1,5,10,1])
dist3= distance.euclidean(E1,E3)
dist3
```

Chebyshev method


```
from scipy.spatial import distance
E1=np.array([10000,1,0,0,0,0,7,15,1])
E3=np.array([15000,1,1,2,1,1,5,10,1])
dist3= distance.chebyshev(E1,E3)
dist3
```

Manhattan Distance

```
from scipy.spatial import distance
E1=np.array([10000,1,0,0,0,0,7,15,1])
E3=np.array([15000,1,1,2,1,1,5,10,1])
dist3= distance.cityblock(E1,E3)
dist3
```

#### Minkowski method

```
from scipy.spatial import distance
E1=np.array([10000,1,0,0,0,0,7,15,1])
E3=np.array([15000,1,1,2,1,1,5,10,1])
dist3= distance.minkowski(E1,E3)
dist3
```
