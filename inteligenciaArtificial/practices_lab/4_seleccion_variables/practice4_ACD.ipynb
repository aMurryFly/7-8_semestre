## **3. Visualization of correlations**

```
plt.plot(Datos['Radius'],Datos['Perimeter'],'ro')
plt.xlabel('Radius')
plt.ylabel('Perimeter')
```

**3.1 Análisis**

Existe tendencia lineal entre las 2 variables, es decir, mientras máás grande sea el radio del tumor a su vez el períímetro del cáncer

-> La correlación es de 0.99

```
plt.plot(Datos['Radius'],Datos['Area'],'ro')
plt.xlabel('Area')
plt.ylabel('Perimeter')
```

**3.2 Análisis**

Existe tendencia lineal entre las 2 variables, es decir, mientras más grande sea el radio del tumor a su vez el área del cáncer también.

-> La correlación es de 0.99


```
plt.plot(Datos['ConcavePoints'],Datos['Concavity'],'ro')
plt.xlabel('ConcavePoints')
plt.ylabel('Concavity')
```

**3.3 Análisis**

Existe una relación entre ambas variables, lo cual nos sugiere que la concavidad y los puntos concavos dentro del cáncer

-> La correlación es de 0.939888


```
plt.plot(Datos['ConcavePoints'],Datos['Compactness'],'ro')
plt.xlabel('ConcavePoints')
plt.ylabel('Compactness')
```

**3.4 Análisis**

Hay una relación bastante buena entre la concavidad de los puntos y lo compacto de los tumores.
